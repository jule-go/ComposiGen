### List of compounds we want to be able to get embeddings for

Idea: look at related works and retrieve a set of noun-noun compounds. Later when training a word2vec model preprocess the corpus in such way that it also learns to capture the meaning of those compounds.


Load my compounds:

```python
path_to_this_repo = "/path/to/repo/" # TODO adapt if needed
my_compounds = set()
with open(path_to_this_repo+"Data/CompositionalityRatings/compounds.csv","r") as infile:
    for line in infile:
        if not line.startswith("ID"): # exclude the header file
            my_compounds.add(line.split(",")[1])
print(len(my_compounds),my_compounds)
```

Load Kurtyigit et al. (2025)'s compounds: (as they are based on the list from Reddy, simply use the Reddy compounds)

```python
Reddy_compounds = set()
with open("/path/to/data/kurtyigit_compounds.txt","r") as infile:
    for line in infile:
        if not line.startswith("#word"): # extract header line
            word = line.split("\t")[0]
            if word.count("-n") == 2: # both constituents are labeled as nouns
                Reddy_compounds.add(word.replace("-n","").strip())
print(len(Reddy_compounds),Reddy_compounds)
```

Load more compounds:

```python
Urs_compounds = set()
with open("/path/to/data/other_compounds.txt","r") as infile:
    for line in infile:
        Urs_compounds.add(line.strip())
print(len(Urs_compounds),Urs_compounds)
```

Load Cordeiro's compounds:

```python
Cordeiro_compounds = set()
with open("/path/to/data/cordeiro_compounds.txt","r") as infile:
    for line in infile:
        if not line.startswith("Compound"): # exclude header line
            Cordeiro_compounds.add(line.split("\t")[0])
print(len(Cordeiro_compounds),Cordeiro_compounds)
```

Merge compounds from related work into one file:

```python
all_compounds = my_compounds.union(Reddy_compounds,Urs_compounds,Cordeiro_compounds)
print(len(all_compounds),all_compounds)

# save to file
with open("targets.txt","w") as outfile:
    for comp in sorted(all_compounds):
        outfile.write(comp+"\n")
```
